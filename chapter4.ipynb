# 4. Curse of dimensionality

## a)

For one dimension, using 10% of the uniformly distributed observations closest to a test item will result in a proportion of $0.1$ of the observations being used.

## b) 

For two dimensions, you'd only have $0.1 ^ 2 = 0.01$ or $\frac{1}{100}$ of the observations on average.

## c) 

Here you're really screwed: you only get $0.1 ^ 100 = 10^{-100}$ of all the observations on average.  You'd need 10 billion observations before you'd have even **one** observation to go by on average.

## d) 

The argument rests on the fact that when there are a lot of features, there are a lot of ways that observations can differ from each other, and it's not always clear which of these many possible differences are relevant for determining the response.

## e) 

If $s$ is the size (proportion) of each dimension that we use, $p$ is the number of dimensions, and $f$ is the fraction of the total number of observations, than like we've seen above, $f = s ^ p$.  If we want to find the required size for a particular proportion, we just invert the power: $s = f ^ {1/p}$.  For $p = 1, 2, 100$ this works out to: 

```julia
[p => 0.1 ^ (1/p) for p in [1, 2, 100]]
```

That is, we'd need to consider **98% of each dimension** to get just **10%** of the observations when we have 100 features (not that many!).  That is, by volume, we're excluding 90% of the space, but on each dimension, we're only excluding like 2%!  Each individual feature becomes basically useless as a dignostic

# 10. `Weekly`

```julia
using DataFrames, Gadfly, RDatasets, DA, GLM, Distributions
# DA is a package for doing discriminant analysis.  It's not registered but you 
# can install it by doing
#Pkg.clone("https://github.com/trthatcher/DA.jl")

weekly = dataset("ISLR", "Weekly")
weekly[:DirectionUp] = [dir == "Up" ? 1 : 0 for dir in weekly[:Direction]]
```

## a) Graphical summaries etc.

```julia
describe(weekly)
```

```julia
#by(, :Direction, mean)
updown_m = melt(weekly, :Direction)
@show updown_summ = by(updown_m, [:variable, :Direction],
                       df -> DataFrame(mean=mean(df[:value]), std=std(df[:value])))
@show updown_std = unstack(updown_summ, :variable, :Direction, :std)
updown = unstack(updown_summ, :variable, :Direction, :mean)
updown[:diff] = updown[:Down] - updown[:Up]
updown[:std] = sqrt((updown_std[:Down].^2 + updown_std[:Up].^2) / 2)
updown[:z] = updown[:diff] ./ updown[:std]
updown
```

## c) Confusion matrix

```julia
direction_up_glm = DataFrame(truth=weekly[:DirectionUp], 
                             pred=convert(Vector{Int}, (predict(weekly_glm) .> 0.5)))
by(direction_up_glm, [:truth, :pred], df -> size(df, 1))
```

It's classifying pretty much everything at "Up", indiscriminately.    

## d) Better GLM?

Training on just 1990 to 2008, testing on 2009, 2010.  Only using `Lag2` as a predictor

```julia
weekly_train = weekly[weekly[:Year] .<= 2008, :]
weekly_test  = weekly[weekly[:Year] .> 2008, :]

weekly_lag2_glm = glm(DirectionUp ~ Lag2, weekly_train, Binomial())
display(weekly_lag2_glm)

pred_logistic(mm::GLM.GeneralizedLinearModel, newX::AbstractArray) = logistic(newX * coef(mm))
pred_logistic(mm::DataFrames.DataFrameRegressionModel, newX::AbstractArray) = pred_logistic(mm.model, newX)
function pred_logistic(mm::DataFrames.DataFrameRegressionModel, df::DataFrame)    
    newX = ModelMatrix(ModelFrame(df, mm.mf.terms, falses(size(df[1]))))
    pred_logistic(mm, newX.m)
end

# if I wanted to do something more general, could extend ModelMatrix to construct 
# from existing ModelFrame/DataFrameRegressionModel with something like
#ModelMatrix(ModelFrame(weekly_test, weekly_lag2_glm.mf.terms, isna(weekly_test[1])))
    

weekly_test[:DirUpPredGLM] = convert(Vector{Int}, pred_logistic(weekly_lag2_glm, weekly_test) .> 0.5)
@show acc_glm = mean(weekly_test[:DirUpPredGLM] .== weekly_test[:DirectionUp])
by(weekly_test, [:DirectionUp, :DirUpPredGLM], df -> size(df, 1))
```

```julia
mm = ModelMatrix(ModelFrame(weekly_test, weekly_lag2_glm.mf.terms, isna(weekly_test[1])))
dump(mm)
```

## e) LDA

```julia
weekly_lda = lda(DirectionUp ~ Lag2, weekly_train)
#dump(weekly_lda)
weekly_test[:DirUpPredLDA] = DA.predict(weekly_lda, weekly_test)[:,1]
@show acc_lda = mean(weekly_test[:DirUpPredLDA] .== weekly_test[:DirectionUp])
by(weekly_test, [:DirectionUp, :DirUpPredLDA], df -> size(df,1))
```

## f) QDA

```julia
weekly_qda = qda(DirectionUp ~ Lag2, weekly_train)
#dump(weekly_lda)
weekly_test[:DirUpPredQDA] = DA.predict(weekly_qda, weekly_test)[:,1]
@show acc_qda = mean(weekly_test[:DirUpPredQDA] .== weekly_test[:DirectionUp])
by(weekly_test, [:DirectionUp, :DirUpPredQDA], df -> size(df,1))
```

## g) KNN

```julia
using kNN, Distance

weekly_knn = knn(array(weekly_train[:Lag2])', array(weekly_train[:DirectionUp]))
weekly_test[:DirUpPredKnn1] = predict(weekly_knn, array(weekly_test[:Lag2])', 1)
@show acc_knn = mean(weekly_test[:DirUpPredKnn1] .== weekly_test[:DirectionUp])
@show by(weekly_test, [:DirectionUp, :DirUpPredKnn1], df -> size(df,1))
```

```julia
knn_acc = k -> mean(weekly_test[:DirectionUp] .== predict(weekly_knn, array(weekly_test[:Lag2])', k))
[(k, knn_acc(k)) for k in [1:2:20, 24:4:40, 50:10:100]]
```
