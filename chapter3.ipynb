# 8. Linear regression on `Auto`

## a) `mpg` vs. `horsepower`

First load the data and fit the model.

```julia
auto = dataset("ISLR", "Auto")
lm_mpg_hp = fit(LinearModel, mpg ~ horsepower, auto)
```

There is definitely a relationship, and it's statistically highly significant ($p < 10^{-80}$).  The coefficient is negative, which means that higher horsepower corresponds to lower mpg.

Here is two ways to predict the expected mpg for a 98 horsepower car:

```julia
newx = [1 98]
newx * coef(lm_mpg_hp)
```

```julia
newx_pred = predict(lm_mpg_hp, newx)
```

```julia
conf_se = sqrt(newx * vcov(lm_mpg_hp) * newx')
p = 0.95
conf_int = quantile(TDist(df_residual(lm_mpg_hp.model)), (1-p)/2) * [1 -1] .* conf_se .+ newx_pred
```

```julia
s = sqrt(sum(residuals(lm_mpg_hp) .^ 2) / df_residual(lm_mpg_hp.model))
total_se = sqrt(conf_se^2 + s^2)
pred_int = quantile(TDist(df_residual(lm_mpg_hp.model)), (1-p)/2) * [1 -1] .* total_se .+ newx_pred
```

## b) Plotting the regression line and data

```julia
# Gadfly can plot a function that takes one number (the x coordinate)
# and returns one number (the y coordinate)
pred = function(x, mod) 
     predict(mod, hcat(1, x))[1]
end

# this is a bit of a hack.  There's a nicer way to plot a function, 
# which is layer(f::Function, a, b), but internally that expands to this
# (which sets y to be an array of functions, xmin=[a], xmax=[b], and
# uses Stat.func and Geom.line).  Unfortunately, that doesn't appear
# to allow for other things to be added (Geoms or Themes).
plot(layer(auto, x=:horsepower, y=:mpg, Geom.point), 
     layer(y=[x->pred(x, lm_mpg_hp)], xmin=[25], xmax=[240], Stat.func, Geom.line, Theme(default_color=color("red"))))
```

# c) Diagnostics (a la `plot.lm` in R)

Residuals vs. fitted (below) reveals the non-linearity of the data: predictions undershoot the data at extreme values, and overshoot (a bit) in the middle.

```julia
resids = auto[[:horsepower, :mpg]]
resids[:mpg_fitted] = predict(lm_mpg_hp)
resids[:mpg_residual] = residuals(lm_mpg_hp)
plot(resids, x=:mpg_fitted, y=:mpg_residual, 
     layer(Geom.point), 
     layer(Geom.smooth, Theme(default_color=color("red"))), 
     Guide.xlabel("Fitted mpg"), 
     Guide.ylabel("Residual mgp"))
```

```julia
# QQ plot, take two
using Distributions

set_default_plot_size(10cm, 10cm)
s_hat = sqrt(sum(resids[:mpg_residual].^2) / df_residual(lm_mpg_hp.model))
plot(layer(resids, x=:mpg_residual, y=Normal(0, s_hat), Geom.point, Stat.qq, Theme(highlight_width=0px)), 
     layer(y=[x -> x], xmin=[-18], xmax=[18], Geom.line, Stat.func, Theme(default_color=color("red"))))
```

# 10. `Carseats`

## a) Fit model

```julia
carseats = dataset("ISLR", "Carseats")
cs_full_lm = fit(LinearModel, Sales ~ Price + Urban + US, carseats)
```

## b) Interpretation

Higher site-specific `price` results in lower sales, while stores that are in the `US` have higher sales.  Stores that are in `Urban` areas have numerically lower sales, but this effect is really basically zero.

## c) Equation form

$y = \beta_0 + \beta_{\mathrm{Price}} x_\mathrm{Price} + \beta_\mathrm{Urban} I(x_\mathrm{Urban}=\mathrm{"Yes"}) + \beta_\mathrm{US} I(x_\mathrm{US}=\mathrm{"Yes"})$

## d) Significant predictors

All but `Urban`.

## e) Reduced model

```julia
cs_red_lm = fit(LinearModel, Sales ~ Price + US, carseats)
```

## f) Compare fit

```julia
r2(m::LinearModel) = 1 - sum((m.rr.y - m.rr.mu).^2) / sum((m.rr.y-mean(m.rr.y)).^2)

@show r2(cs_full_lm.model)
@show r2(cs_red_lm.model)
```

## g) Confidence intervals

```julia
confint(cs_red_lm)
```

## h) Outliers/leverage

```julia
plot(x=predict(cs_red_lm), y=residuals(cs_red_lm), Geom.point)
```

# 14. Collinearity

## a) Generate data

```julia
srand(1)
x1 = rand(Normal(), 100)
x2 = 0.5 .* x1 + rand(Normal(0., 0.1), 100)
y = 2 + 2 .* x1 + 0.3 .* x2 + rand(Normal(), 100)
```

The regression coefficients are $\beta_0 = 2$, $\beta_1 = 2$, and $\beta_2 = 0.3$

## b) Correlation/scatterplot

```julia
@show cor(x1, x2)
plot(x=x1, y=x2, Geom.point)
```

## c) Fit linear model

```julia
lm_both = fit(LinearModel, hcat(x1, x2), y)
```

## d) Just `x1`

```julia
lm_x1 = fit(LinearModel, hcat(x1), y)
```

## e) Just `x2`

```julia
lm_x2 = fit(LinearModel, hcat(x2), y)
```

## f) Contradiction? 

No, they don't.  Both `x1` and `x2` predict `y` well, but they are redundant.  If only one is availabile, it is a good predictor, but when taken together, `x1` is better ($\beta_1 = 2$) than `x2` ($\beta_2 = 0.3$).

## g) Additional observation

```julia
push!(x1, 0.1)
push!(x2, 0.8)
push!(y, 6)
new_point = fill(false, 100)
push!(new_point, true)
```

```julia
plot(x=x1, y=y)
```

```julia
plot(x=x2, y=y)
```

```julia
@show lm_both = fit(LinearModel, hcat(x1, x2), y)
@show lm_x1 = fit(LinearModel, hcat(x1), y)
@show lm_x2 = fit(LinearModel, hcat(x2), y)
```

```julia
plot(x=predict(lm_both), y=residuals(lm_both), color=new_point)
```

```julia
plot(x=predict(lm_x1), y=residuals(lm_x1), color=new_point)
```

```julia
plot(x=predict(lm_x2), y=residuals(lm_x2), color=new_point)
```

It doesn't look like either an outlier (except _maybe_ for the `x1` model) or a high-leverage point.
